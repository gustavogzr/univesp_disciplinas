```python
from nltk.tokenize import TweetTokenizer # importar o tokenizador de tweets
texto2 = "I'm very veryyyy happyyyyyyyy #betterlife @barneys :P :D"
```

```python
word_tokenize(texto2) # tokenizar o texto2 considerando somente as palavras
```

Veja que o tokenize "normal" não funciona adequadamente para tweets.

```python
twitterTokenizer = TweetTokenizer() # instanciar o tokenizador de tweets
twitterTokenizer.tokenize(texto2) # tokenizar o texto2 considerando somente as palavras
```

```python
twitterTokenizer2 = TweetTokenizer(
    strip_handles=True, # remover os handles, ou seja, os @
    reduce_len=True # reduzir o tamanho das palavras repetidas, para no máximo 3 caracteres
 )
twitterTokenizer2.tokenize(texto2) # tokenizar o texto2 considerando somente as palavras
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')

print(nltk.ne_chunk( # ne_chunk é usado para identificar entidades nomeadas
    nltk.pos_tag( # pos_tag é usado para identificar a classe gramatical das palavras
        nltk.word_tokenize(text) # word_tokenize é usado para tokenizar o texto
    )))
```

```python
# fazer download do pacote de dados do spacy (en_core_web_sm)
!python -m spacy download en_core_web_sm
```

```python
import spacy
nlp = spacy.load('en_core_web_sm') # carregar o modelo em inglês
sentenca = nlp('Lucas is working at Google in the South America. He works very hard!')
for token in sentenca:
    print(token.text, token.pos_, token.dep_, token.tag_) # exibir o texto, a classe gramatical, a dependência sintática e a tag
```

```python
doc = nlp('Lucas is working at Google in the South America. He works very hard!')
for sent in doc.sents:
    print(sent.text) # exibir as sentenças
```

```python
sentenca = nlp('Lucas is working at Google in the South America.\
               He works very hard!')
for token in sentenca:
    print(f'{token.text} - {token.is_stop}') # exibir o texto e se a palavra é uma stopword
```

```python
sentenca = nlp('Lucas is working at Google in the South America since 1999.\
               He works very hard! This product cost $90.89')
for token in sentenca:
    print(f'{token.text} - {token.is_alpha}') # exibir o texto e se a palavra é alfabética
```

```python
setenca = nlp('Lucas has drinked two coffees while the\
              computer is conputing the values of the matrices.')
for token in setenca:
    print(f'{token.text} - {token.lemma_}') # exibir o texto e o lema da palavra
```
