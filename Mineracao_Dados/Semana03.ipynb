```python
# importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importar dados do arquivo 'Birthweight_reduced_kg_R.csv'
df_birth = pd.read_csv('Birthweight_reduced_kg_R.csv')
df_birth.head()
```

```python
# Fazer histograma com a variável "Birthweight"
sns.displot(df_birth['Birthweight'])
plt.show()
```

```python
# Fazer histograma com a variável "Length"
sns.displot(df_birth['Length'])
plt.show()
```

```python
# Construir gráfico de setores das mães que fumam (smoker)
num_fumantes = len(df_birth[df_birth['smoker'] == 1])
num_nao_fumantes = len(df_birth[df_birth['smoker'] == 0])

labels = ['Fumantes', 'Não Fumantes']
plt.pie([num_fumantes, num_nao_fumantes], labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# Obter a média da variável "length"
media_altura = df_birth['Length'].mean()
print("Média da altura dos bebês =", media_altura)
# Obter o desvio-padrão da variável "length"
desvio_altura = df_birth['Length'].std()
print("Desvio-padrão da altura dos bebês =", desvio_altura)
```

```python
# Construir um diagrama de caixa para a idade da mãe (mage) com legendas numéricas
sns.boxplot(df_birth['mage'])
plt.show()
```

```python
# Construir gráfico de dispersão entre o peso do bebê (Birthweight) e tempo de gestação (Gestation)
sns.scatterplot(x=df_birth['Birthweight'], y=df_birth['Gestation'], hue=df_birth['smoker'])
plt.grid()
plt.xlabel('Peso do bebê')
plt.ylabel('Tempo de gestação')
plt.show()
```

```python
# Reconstruir gráfico anterior com segregação de mães que fumam e que não fumam (smoker)
plt.scatter(df_birth['Birthweight'], df_birth['Gestation'], c=df_birth['smoker'])
plt.grid()
plt.xlabel('Peso do bebê')
plt.ylabel('Tempo de gestação')
plt.show()
```

```python
# importar arquivos "Quantidade de alunos por tipo de ensino da rede estadual_2021_1°SEMESTRE.csv"
df_escolas = pd.read_csv('Quantidade de alunos por tipo de ensino da rede estadual_2021_1°SEMESTRE.csv', sep=';')
df_escolas.head()
```

```python
df_escolas.info()
```

```python
sns.displot(df_escolas['ANOS INICIAIS'])
plt.show()
```

```python
filtrar_escolas = df_escolas[df_escolas['ANOS INICIAIS'] > 0]
sns.displot(filtrar_escolas['ANOS INICIAIS'], bins=10)
plt.show()
```

```python
filtrar_escolas = df_escolas[df_escolas['ANOS FINAIS'] > 0]
sns.displot(filtrar_escolas['ANOS FINAIS'], bins=10)
plt.show()
```

```python
filtrar_escolas = df_escolas[df_escolas['ENSINO MÉDIO'] > 0]
sns.displot(filtrar_escolas['ENSINO MÉDIO'], bins=10)
plt.show()
```

```python
num_fundamental_iniciais = df_escolas['ANOS INICIAIS'].sum()
num_fundamental_finais = df_escolas['ANOS FINAIS'].sum()
num_medio = df_escolas['ENSINO MÉDIO'].sum()
(num_fundamental_iniciais + num_fundamental_finais + num_medio)
```

```python
alunos = [num_fundamental_iniciais, num_fundamental_finais, num_medio]
labels = ['Fundamental Iniciais', 'Fundamental Finais', 'Ensino Médio']
plt.pie(alunos, labels=labels, autopct='%0.1f%%')
plt.show()
```

```python
filtrar_escolas = df_escolas[df_escolas['ENSINO MÉDIO'] > 0]
sns.boxplot(y = filtrar_escolas['ENSINO MÉDIO'])
plt.show()
```

```python
df_escolas['ANOS FINAIS'].describe()
```

```python
filtrar_escolas = df_escolas[(df_escolas['ANOS INICIAIS'] > 0) & (df_escolas['ANOS FINAIS'] > 0)]
sns.scatterplot(x = filtrar_escolas['ANOS INICIAIS'], y = filtrar_escolas['ANOS FINAIS'])
plt.show()
print('Correlação entre ANOS INICIAIS e ANOS FINAIS =', filtrar_escolas['ANOS INICIAIS'].corr(filtrar_escolas['ANOS FINAIS']))
```
