```python
# importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importar dados do arquivo 'Birthweight_reduced_kg_R.csv'
df_birth = pd.read_csv('Birthweight_reduced_kg_R.csv')
df_birth.head()
```

```python
# Fazer histograma com a variável "Birthweight"
plt.hist(df_birth['Birthweight'])
plt.show()
```

```python
# Fazer histograma com a variável "Length"
plt.hist(df_birth['Length'])
plt.show()
```

```python
# Construir gráfico de setores das mães que fumam (smoker)
df_birth['smoker'].value_counts().plot.pie()
```

```python
# Obter a média da variável "length"
media_altura = df_birth['Length'].mean()
print("Média da altura dos bebês =", media_altura)
# Obter o desvio-padrão da variável "length"
desvio_altura = df_birth['Length'].std()
print("Desvio-padrão da altura dos bebês =", desvio_altura)
```

```python
# Construir um diagrama de caixa para a idade da mãe (mage) com legendas numéricas
plt.boxplot(df_birth['mage'], labels=['Idade da mãe'])
plt.show()
```

```python
# Construir gráfico de dispersão entre o peso do bebê (Birthweight) e tempo de gestação (Gestation)
plt.scatter(df_birth['Birthweight'], df_birth['Gestation'])
plt.grid()
plt.xlabel('Peso do bebê')
plt.ylabel('Tempo de gestação')
plt.show()
```

```python
# Reconstruir gráfico anterior com segregação de mães que fumam e que não fumam (smoker)
plt.scatter(df_birth['Birthweight'], df_birth['Gestation'], c=df_birth['smoker'])
plt.grid()
plt.xlabel('Peso do bebê')
plt.ylabel('Tempo de gestação')
plt.show()
```
