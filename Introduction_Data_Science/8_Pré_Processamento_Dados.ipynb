```python
dados2 = pd.read_csv("AAPL.csv")
dados2
```

```python
dados2['close_ontem'] = dados2['Close'].shift(1)
dados2
```

```python
dados2['alteração %'] = ((dados2['Close']/dados2['close_ontem'])-1)*100
dados2.head(20)
```
