### 5 – Qual é a nota mais atribuída? Quantas vezes ela foi atribuída? Quantos % das notas totais?


```python
groupby = df_apple.groupby('Ratings')['Ratings'].count()
df_groupby = pd.DataFrame({'Rating':groupby.index, 'QTDE':groupby.values})
QTDE_TT = df_groupby['QTDE'].sum()
df_groupby['Porcentagem'] = 100*df_groupby['QTDE']/QTDE_TT
df_groupby.sort_values(by='Porcentagem', ascending=False)
```

**Resposta:** Rating 5 foi atribuído 6.788 vezes e representa 69,9% das notas totais
