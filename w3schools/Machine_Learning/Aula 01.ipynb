```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
x = np.mean(speed)
print(x)
```

```python
x = np.median(speed)
print(x)
```

```python
x = stats.mode(speed)
print(x)
```

https://www.w3schools.com/python/python_ml_linear_regression.asp

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y) # regressão linear

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x)) # mapeia os valores de x para a função myfunc

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()

print(f'Coeficiente de correlação: {r}\n') # coeficiente de correlação

speed_10 = myfunc(10) # velocidade prevista para um carro com 10 anos

print(f'Velocidade prevista para um carro com 10 anos: {speed_10}')
```

```python
# Regressão linear com baixa correlação

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y) # regressão linear

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x)) # mapeia os valores de x para a função myfunc

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()

print(f'Coeficiente de correlação: {r}\n') # coeficiente de correlação
```
