```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importar arquivo de url

df_sorvete = pd.read_csv('https://raw.githubusercontent.com/higoramario/univesp-com400-visualizacao-computacional/main/semana-2-exercicio-apoio-1-vendas-sorvete.csv')
df_sorvete.head()
```

```python
df_sorvete.describe()
```

```python
# Calcular a correlação:
correlacao = df_sorvete['temperatura'].corr(df_sorvete['vendas'])

# Criar gráfico de dispersão:
plt.scatter(df_sorvete['temperatura'], df_sorvete['vendas'], color='red')
plt.title('Temperatura vs Vendas')
plt.xlabel('Temperatura')
plt.ylabel('Vendas')

plt.text(10, 500, f'Correlação: {correlacao:.2f}') # incluir correlação no gráfico

plt.show()
```

```python
# Importar novo arquivo de url
df_campeoes = pd.read_csv('https://raw.githubusercontent.com/higoramario/univesp-com400-visualizacao-computacional/main/semana-2-exercicio-apoio-2-campeoes-mundiais.csv')
df_campeoes.head()
```

```python
df_campeoes_agrupado = df_campeoes.groupby('CAMPEÃO').size().reset_index(name='QTDE de Títulos').sort_values(by='QTDE de Títulos', ascending=False)

plt.bar(df_campeoes_agrupado['CAMPEÃO'], df_campeoes_agrupado['QTDE de Títulos'])
plt.title('Títulos Mundiais por Time')
plt.ylabel('Quantidade de Títulos')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.gcf().set_size_inches(10, 5)
plt.show()
```

```python
# construir gráfico de barras horizontais com os maiores campeões na parte superior
df_campeoes_agrupado = df_campeoes_agrupado.sort_values(by='QTDE de Títulos', ascending=True)
plt.barh(df_campeoes_agrupado['CAMPEÃO'], df_campeoes_agrupado['QTDE de Títulos'])
plt.title('Títulos Mundiais por Time')
plt.xlabel('Quantidade de Títulos')
plt.grid(axis='x')
plt.gcf().set_size_inches(10, 10)
plt.show()
```
