```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# importar dados
url_dados = 'https://raw.githubusercontent.com/higoramario/univesp-com400-visualizacao-computacional/main/com400-semana-7-disciplinas-notas.csv'
df_notas = pd.read_csv(url_dados)
df_notas.head()
```

```python
df_notas.describe()
```

- A lista possui 500 notas
- A média das notas de COM400 é 4,9
- A média das notas de COM310 é 5,11
- A média de idade é 40,9 anos, com idade mínima de 18 anos e máxima de 65 anos

```python
# criar histograma de idades com seaborn com 47 bins
sns.histplot(df_notas['idade'], bins=47)
plt.xlabel('Idade')
plt.ylabel('Frequência Absoluta')
plt.title('Histograma de Idades')
plt.show()
```

```python
df_notas['sexo'] = df_notas['sexo'].map({'F': 'Feminino', 'M': 'Masculino'})

sns.countplot(data=df_notas, x='sexo', palette='colorblind')
plt.xlabel('Sexo')
plt.ylabel('Frequência Absoluta')
plt.title('Quantidade de Estudantes por Sexo')
plt.show()
```

```python
sns.violinplot(data=df_notas[['com400-nota', 'com310-nota']], palette='colorblind').set_xticklabels(['COM310','COM400'])
plt.xlabel('Disciplina')
plt.ylabel('Nota')
plt.title('Distribuição das Notas das Disciplinas')
plt.show()
```

```python
# Gere um gráfico de dispersão comparando a idade dos estudantes e as notas da disciplina COM400. Faça o gráfico de dispersão do tipo kde usando a função jointplot().
sns.jointplot(data=df_notas, x='idade', y='com400-nota', kind='kde', color='blue')
plt.xlabel('Idade')
plt.ylabel('Nota na Disciplina COM400')
plt.title('Distribuição das Notas da Disciplina COM400 por Idade')
plt.show()
```
