```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Importar dados:
df_precos = pd.read_csv('https://raw.githubusercontent.com/higoramario/univesp-com400-visualizacao-computacional/main/semana-4-exercicio-apoio-precos-alimentos.csv')
df_precos.head(10)
```

```python
# Criar gráfico de linha para os preços dos alimentos
plt.figure(figsize=(10,6))
for row in df_precos.iterrows():
    if row[0] < 8:
        plt.plot(df_precos.columns[1:], row[1][1:], label=row[1][0])
    else:
        plt.plot(df_precos.columns[1:], row[1][1:], label=row[1][0], linestyle='--')
# criar legenda em 2 colunas
plt.legend(ncol=2)
plt.title('Preços dos alimentos')
plt.xlabel('Ano')
plt.ylabel('Preço')
plt.show()
```

```python
# criar dataframe "melted"
df_melted = df_precos.melt(id_vars=['Produto'], var_name='Ano', value_name='Preço')
```

```python
plt.figure(figsize=(14, 6))
custom_palette = ['#A5E600', '#E6DE00', '#E5C000', '#E59F00', '#E68200']
sns.barplot(x='Produto', y='Preço', hue='Ano', data=df_melted, palette=custom_palette)
plt.title('Preços dos alimentos')
plt.xlabel('')
plt.xticks(rotation=90)
plt.show()
```
